```python
import numpy as np
from IPython.display import set_matplotlib_formats
import matplotlib.cm as cm 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from scipy.io import savemat
import numpy as np

%matplotlib inline
set_matplotlib_formats('png')

anjie_theme = {'axes.grid': True,
              'grid.linestyle': '--',
              'legend.framealpha': 1,
              'legend.facecolor': 'white',
              'legend.shadow': True,
              'legend.fontsize': 14,
              'legend.title_fontsize': 16,
              'xtick.labelsize': 14,
              'ytick.labelsize': 14,
              'axes.labelsize': 16,
              'axes.titlesize': 20,
              'figure.dpi': 100}
 
matplotlib.rcParams.update(anjie_theme)
```

```python
thres=np.arange(0,0.26,0.01)
def gen_thres_list(data, thres=np.arange(0,0.26,0.001)):
    dis = []
    for i in range(len(thres)):
        temp_num = np.where(data<thres[i])[0]
        dis.append(temp_num.shape[0]/data.shape[0])
    return dis

def get_data(f_path):
    return np.load(f_path, allow_pickle=True)[None][0]['by_uv_dist']
```

```python
dis_implicit = gen_thres_list(get_data("../results/airplane/implicit.npy"), thres)
dis_ours = gen_thres_list(get_data("../results/airplane/ours.npy"), thres)
dis_fold = gen_thres_list(get_data("../results/airplane/foldingnet.npy"), thres)
dis_atlas = gen_thres_list(get_data("../results/airplane/atlasnet.npy"), thres)


sns.lineplot(x=thres, y=dis_implicit, label='Liu et al.')
sns.lineplot(x=thres, y=dis_atlas, label='AtlasNetV2')
sns.lineplot(x=thres, y=dis_fold, label='FoldingNet')
sns.lineplot(x=thres, y=dis_ours, label='Ours')
plt.title('Airplane')
```
